```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
t = np.linspace(0, 2 * np.pi, 1000)
a1, a2 = 1.5, 1.0
phases = [0, np.pi/4, np.pi/2]
labels = ['0', r'\pi/4', r'\pi/2', r'\pi']
colors = ['green', 'blue', 'red']

fig, ax = plt.subplots(1, 1, figsize=(8, 8))

for i, delta in enumerate(phases):
    x = a1 * np.cos(t)
    y = a2 * np.cos(t + delta)

    ax.plot(x, y, color=colors[i], lw=2, label=f'$\delta = {labels[i]}$')

plt.gca().add_patch(plt.Rectangle((-a1, -a2), 2*a1, 2*a2,
                                   fill=False, color='black', lw=2, label='Bounding Box'))
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel('y', fontsize=16)
ax.legend(fontsize=14, loc=2)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2 * np.pi, 1000)
a1, a2 = 2.0, 1.2
delta = np.pi/3

x = a1 * np.cos(t)
y = a2 * np.cos(t + delta)

plt.figure(figsize=(7, 6))
plt.plot(x, y, 'blue', lw=2, label='Locus of P')

plt.gca().add_patch(plt.Rectangle((-a1, -a2), 2*a1, 2*a2,
                                   fill=False, color='red', ls='--', lw=1.5, label='Bounding Box'))

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.legend(fontsize=14)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

ratios = [(1, 2), (1, 3)]
a1, a2 = 1.5, 1.0

fig, ax = plt.subplots(1, 1, figsize=(8, 8))

colors=['red', 'black']

for i, (wx, wy) in enumerate(ratios):
    t = np.linspace(0, 2 * np.pi, 1000)
    x = a1*np.cos(wx * t)
    y = a2*np.cos(wy * t + delta)

    ax.plot(x, y, lw=1.5, label=f'Ratio $\omega_x:\omega_y = {wx}:{wy}$', color=colors[i])

ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_xlabel('x', fontsize=16)
ax.set_ylabel('y', fontsize=16)
ax.legend(loc='upper right', fontsize=14)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

ax.set_xlim(-1.75, 1.75)
ax.set_ylim(-1.1, 1.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

omega_x = 1.0
omega_y = np.pi
t_max = 5000
num_points = 2000000

t = np.linspace(0, t_max, num_points)

x = 1.5*np.cos(omega_x * t)
y = np.cos(omega_y * t)

bins = [30, 20]
heatmap, xedges, yedges = np.histogram2d(x, y, bins=bins, range=[[-1.5, 1.5], [-1.0, 1.0]])

non_zero_heatmap_values = heatmap[heatmap > 0]
if non_zero_heatmap_values.size > 0:
    vmax_val = np.percentile(non_zero_heatmap_values, 99)
else:
    vmax_val = None

plt.imshow(
    heatmap.T,
    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
    origin='lower',
    cmap='magma',
    interpolation='nearest',
    vmax=vmax_val
)
cbar = plt.colorbar(label='Visitation Frequency')
cbar.set_label('Visitation Frequency', fontsize=16)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.savefig('oscillator_heatmap.png', bbox_inches='tight')
```
